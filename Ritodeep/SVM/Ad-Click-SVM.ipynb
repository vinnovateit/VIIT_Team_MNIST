```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\RITODEEP\Downloads\dataset\train_.csv',encoding='latin-1')
```

```python
print(df.isnull().sum())
df.isnull().sum().sum()
```

```python
df["tstamp"] = pd.to_datetime(df["hour"], format="%y%m%d%H")
df["year"] = df["tstamp"].dt.year
df["month"] = df["tstamp"].dt.month
df["day"] = df["tstamp"].dt.day
df["hours"] = df["tstamp"].dt.hour
df
```

```python
a=df.pop('click')
df['click']=a
```

```python
df.info()
```

```python
df=pd.DataFrame(df,columns=['id', 'hour','site_id',
       'site_domain', 'site_category', 'app_id', 'app_domain', 'app_category', 'device_model', 'device_type',
       'device_conn_type','hours', 'banner_pos', 'C1','C14', 'C15', 'C16','C17', 'C18', 'C19', 'C20', 'click'])
df.info()
```

```python
df.iloc[:,2:12]=df.iloc[:,2:12].astype('category')
```

```python
for i in range(2,12):
    df.iloc[:,i]=df.iloc[:,i].cat.codes
```

```python
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5,metric = 'minkowski',p=2)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred,average='macro')
```

```python
from sklearn.metrics import f1_score
y_t = [0, 1, 1, 0, 1, 1]
y_p = [0, 1, 1, 0, 1, 1]
f1_score(y_t, y_p)  
```

```python
from sklearn.metrics import fbeta_score
fbeta_score(y_test, y_pred,beta=0.5)
```

```python
from sklearn.metrics import log_loss  
# where y_pred are probabilities and y_true are binary class labels
log_loss(y_test, y_pred, eps=1e-15)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, y_pred, average='macro')
```
